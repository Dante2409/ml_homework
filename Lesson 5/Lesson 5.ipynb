```python
def f(x):
    return 3 * x ** 2

def df(x):
    return 6 * x
```

```python
x0 = 5
lambda_ = 0.1
threshold = 1e-8
x_history = [x0]

while True:
    x_next = x0 - lambda_ * df(x0)
    if abs(x_next - x0) < threshold:
        break
    x0 = x_next
    x_history.append(x0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_history, [f(x) for x in x_history], '-o')
plt.xlabel('x value')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x, y):
    return 3 * np.square(x) + 5 * np.square(y)

def df(x, y):
    return np.array([6*x, 10 * y])
```

```python
x0, y0 = 5, 10
lambda_ = 0.1
threshold = 1e-8
history = [(x0, y0)]

while True:
    x_next, y_next = (x0 - lambda_ * df(x0, y0)[0], y0 - lambda_ * df(x0, y0)[1])
    if (x_next - x0) ** 2 + (y_next - y0) ** 2 < threshold ** 2:
        break
    x0, y0 = x_next, y_next
    history.append((x0, y0))
```

```python
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
z = f(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6)

history = np.array(history)
z_points = f(history[:, 0], history[:, 1])
ax.plot(history[:, 0], history[:, 1], z_points, 'r-', marker='o')

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
